```python
import requests
from bs4 import BeautifulSoup
import csv

url = 'https://www.healthline.com/nutrition/27-health-and-nutrition-tips'

response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')

    # Targeting the main content of the article
    main_content = soup.find('div', {'class': 'css-1l7w6pa'})

    # Extracting text content from the main content
    if main_content:
        data = main_content.get_text(separator='\n', strip=True)

        # Save data to a text file for better readability
        with open('health_data.txt', 'w', encoding='utf-8') as txt_file:
            txt_file.write(data)

        # Save data to a CSV file (optional)
        csv_file_path = 'health_data.csv'
        with open(csv_file_path, 'w', newline='', encoding='utf-8') as csv_file:
            csv_writer = csv.writer(csv_file)

            # Write header
            csv_writer.writerow(['Health Information'])

            # Write data to CSV
            csv_writer.writerow([data])

        print(f'Data has been scraped and saved to health_data.txt and {csv_file_path}')
    else:
        print('Error: Main content not found on the webpage.')
else:
    print(f'Error: Unable to fetch the webpage. Status Code: {response.status_code}')
```

```python
import requests
from bs4 import BeautifulSoup
import csv

url = 'https://en.wikipedia.org/wiki/Python_(programming_language)'

response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')

    # Extract and print all text data
    all_text_data = soup.get_text(separator='\n', strip=True)
    print(f'All Text Data:\n{all_text_data}\n')

    # Extract and print all HTML tags
    all_html_tags = [tag.name for tag in soup.find_all()]
    print(f'All HTML Tags:\n{all_html_tags}\n')

    # Extract and print data from specific elements (adjust as needed)
    specific_elements_data = [element.get_text(strip=True) for element in soup.select('.mw-content-text p')]
    print(f'Specific Elements Data:\n{specific_elements_data}\n')

    # Save data to CSV (optional)
    csv_file_path = 'wikipedia_python_data.csv'
    with open(csv_file_path, 'w', newline='', encoding='utf-8') as csv_file:
        csv_writer = csv.writer(csv_file)

        # Write header
        csv_writer.writerow(['Data Type', 'Data'])

        # Write data to CSV
        csv_writer.writerow(['All Text Data', all_text_data])
        csv_writer.writerow(['All HTML Tags', ', '.join(all_html_tags)])
        csv_writer.writerow(['Specific Elements Data', ', '.join(specific_elements_data)])

    print(f'Data has been scraped and saved to {csv_file_path}')
else:
    print(f'Error: Unable to fetch the webpage. Status Code: {response.status_code}')
```

```python
import requests
from bs4 import BeautifulSoup

# Replace this URL with the target website's URL
url = 'https://www.healthline.com/nutrition/27-health-and-nutrition-tips'

# Send an HTTP request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content of the page using BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find all visible text elements on the page
    text_elements = soup.find_all(text=True)

    # Create or open a text file to save the scraped data
    with open('scraped_data.txt', 'w', encoding='utf-8') as file:
        # Write the scraped data to the file
        for text in text_elements:
            file.write(text.strip() + '\n')

    print('Scraped data saved to scraped_data.txt')

else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")
```

```python
import requests
from bs4 import BeautifulSoup
import csv

import requests
from bs4 import BeautifulSoup
import csv

# Disable SSL certificate verification
requests.packages.urllib3.disable_warnings(requests.packages.urllib3.exceptions.InsecureRequestWarning)

# Replace this URL with the target website's URL
url = 'https://www.healthline.com/nutrition/27-health-and-nutrition-tips'
# Send an HTTP request to the URL without SSL verification
response = requests.get(url, verify=False)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content of the page using BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find all visible text elements on the page
    text_elements = [text.strip() for text in soup.find_all(text=True) if text.strip()]

    # Create or open a CSV file to save the scraped data
    with open('scraped_data.csv', 'w', encoding='utf-8', newline='') as csvfile:
        # Create a CSV writer
        csv_writer = csv.writer(csvfile)

        # Write header
        csv_writer.writerow(['Text'])

        # Write the scraped data to the file
        for text in text_elements:
            csv_writer.writerow([text])

    print('Scraped data saved to scraped_data.csv')

else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")
```
